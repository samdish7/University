Running a correlation test between each of the different tickers would help us determine if there is a relationship between any of them. For instance, if two tickers have ~1.00 correlation, that means both stocks move up/down in tandem. If two tickers are ~ -1.00, then one stock goes up while the other goes down. If the correlation is close to 0.00, that means the tickers move independently of each other which is considered to be uncorrelated. 

By looking at our line plots above, you can get a visual example and determine whether a ticker is correlated with another. If the lines on different graphs seem to match up then they have a correlation. In this case, the stock seem to be correlated to some degree.
